## 2.4 Target Variable

`Diagnosis` is the label we're predicting: 1 = Alzheimer's diagnosed, 0 = not diagnosed.

```python
target_counts = df[TARGET_COL].value_counts()
target_pct = df[TARGET_COL].value_counts(normalize=True) * 100
print(target_counts)
print()
print(target_pct.round(2))
print(f"\nClass imbalance ratio (majority:minority) ≈ {target_counts.max()/target_counts.min():.2f} : 1")
```

**Observation:** The dataset is moderately imbalanced (~65% No Diagnosis vs ~35% Diagnosis). This is not severe enough to require synthetic resampling (e.g. SMOTE), but it does mean:
- Accuracy alone would be a misleading metric.
- Model comparison in `06_Model_Development.ipynb` uses `class_weight='balanced'` where supported, and reports precision/recall/F1/ROC-AUC rather than accuracy alone.
- Stratified train/val/test splitting is essential (implemented in `src/preprocessing.py::split_data`) so every split preserves this ratio.

## 2.5 Identifier Columns (to be dropped before modelling)

```python
print(df['PatientID'].nunique(), 'unique PatientIDs out of', len(df), 'rows')
print(df['DoctorInCharge'].unique())
```

`PatientID` is a unique row identifier and `DoctorInCharge` is a constant placeholder (`XXXConfid`, a redacted/confidentiality field) — both carry zero predictive signal and are dropped in `src/preprocessing.py::drop_non_predictive_columns`.

## 2.6 Value Range Sanity Check (clinical plausibility)

```python
clinical_ranges = {
    'Age': (60, 90), 'MMSE': (0, 30), 'FunctionalAssessment': (0, 10), 'ADL': (0, 10),
    'SystolicBP': (90, 180), 'DiastolicBP': (60, 120), 'BMI': (15, 45),
}
for col, (lo, hi) in clinical_ranges.items():
    actual_lo, actual_hi = df[col].min(), df[col].max()
    status = "OK" if actual_lo >= lo and actual_hi <= hi else "CHECK"
    print(f"{col:25s} expected [{lo}, {hi}]  actual [{actual_lo:.1f}, {actual_hi:.1f}]  -> {status}")
```

**Summary of Data Understanding:**
- 2,149 patient records, 35 columns, no obvious structural issues at first pass.
- All values fall within clinically plausible ranges.
- Target is moderately imbalanced (~65/35) — informs modelling strategy.
- Two columns (`PatientID`, `DoctorInCharge`) will be dropped as non-predictive.
